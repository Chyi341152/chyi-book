```python
%matplotlib notebook
import matplotlib.font_manager as mfm 
# IPython display module use the HTML() or display()
from IPython.display import display, HTML

font_path = '/usr/share/fonts/custom/STHeitiSC-Light.otf'
prop = mfm.FontProperties(fname=font_path)

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

plt.rcParams['axes.grid'] = True

stockManagerDF = pd.read_csv('./data-set/managerPerformance_stockManager.csv')
mixStockManagerDF = pd.read_csv('./data-set/managerPerformance_mixStockManager.csv')

stockManagerDF['IR'] = stockManagerDF['年化Alpha'] / stockManagerDF['波动率']

#display(stockManagerDF.sort_values(by=['年化Alpha'], ascending=False))

display(stockManagerDF.corr(method='pearson', min_periods=120))
#display(HTML(mixStockManagerDF.to_html()))
```

```python
# 公募基金全能阿尔法之王
print('公募基金股票型基金年化Alpha之王')
display(stockManagerDF.sort_values(by=['年化Alpha'], ascending=False)[['基金经理','年化Alpha']].head(10))

# 公募基金牛市阿尔法之王
print('公募基金经理股票型基金牛市Alpha之王')
display(stockManagerDF.sort_values(by=['牛市Alpha'], ascending=False)[['基金经理','牛市Alpha']].head(10))

# 公募基金熊市阿尔法之王
print('公募基金经理股票型基金熊市Alpha之王')
display(stockManagerDF.sort_values(by=['熊市Alpha'], ascending=False)[['基金经理','熊市Alpha']].head(10))

# 公募基金成长阿尔法之王
print('公募基金经理股票型基金成长Alpha之王')
display(stockManagerDF.sort_values(by=['成长Alpha'], ascending=False)[['基金经理','成长Alpha']].head(10))

# 公募基金成长阿尔法之王
print('公募基金经理股票型基金价值Alpha之王')
display(stockManagerDF.sort_values(by=['价值Alpha'], ascending=False)[['基金经理','价值Alpha']].head(10))

print(''*50)

print('公募基金混合偏股型基金年化Alpha之王')
display(mixStockManagerDF.sort_values(by=['年化Alpha'], ascending=False)[['基金经理','年化Alpha']].head(10))

print('公募基金混合偏股型基金牛化Alpha之王')
display(mixStockManagerDF.sort_values(by=['牛市Alpha'], ascending=False)[['基金经理','牛市Alpha']].head(10))

print('公募基金混合偏股型基金熊化Alpha之王')
display(mixStockManagerDF.sort_values(by=['熊市Alpha'], ascending=False)[['基金经理','熊市Alpha']].head(10))

print('公募基金混合偏股型基金成长Alpha之王')
display(mixStockManagerDF.sort_values(by=['成长Alpha'], ascending=False)[['基金经理','成长Alpha']].head(10))

print('公募基金混合偏股型基金价值Alpha之王')
display(mixStockManagerDF.sort_values(by=['价值Alpha'], ascending=False)[['基金经理','价值Alpha']].head(10))

print(''*50)

print('一年以上5年一下阿尔法之王（每年按照240天计算）-股票型基金')
display(stockManagerDF[(stockManagerDF['总交易日'] >= 240) & (stockManagerDF['总交易日'] < 1200)].sort_values(by=['年化Alpha'], ascending=False)[['基金经理','年化Alpha','总交易日']].head(10))

print('5年以上阿尔法之王（每年按照240天计算）-股票型基金')
display(stockManagerDF[stockManagerDF['总交易日'] >= 1200].sort_values(by=['年化Alpha'], ascending=False)[['基金经理','年化Alpha','总交易日']].head(10))

print('10年以上阿尔法之王（每年按照240天计算）-股票型基金')
display(stockManagerDF[stockManagerDF['总交易日'] >= 2400].sort_values(by=['年化Alpha'], ascending=False)[['基金经理','年化Alpha','总交易日']].head(10))

print('5年以上阿尔法之王（每年按照240天计算）管理规模10亿以上 -股票型基金')
display(stockManagerDF[(stockManagerDF['总交易日'] >= 1200) & (stockManagerDF['规模中间值'] > 1000000000)].sort_values(by=['年化Alpha'], ascending=False)[['基金经理','年化Alpha','总交易日','规模中间值']].head(10))
    
```

```python
# 绘制馅饼图
```

```python
# 选股集中度和业绩的关系
print('选股集中度与Alpha的关系')
DF = stockManagerDF[['牛市Alpha','重仓股集中度']]
DF['重仓股集中度'] = DF['重仓股集中度']/100
DF = DF.dropna(axis=0, how='any')
X1 = DF['重仓股集中度'].tolist()
Y1 = DF['牛市Alpha'].tolist()
del DF

DF = stockManagerDF[['熊市Alpha','重仓股集中度']]
DF['重仓股集中度'] = DF['重仓股集中度']/100
DF = DF.dropna(axis=0, how='any')
X2 = DF['重仓股集中度'].tolist()
Y2 = DF['熊市Alpha'].tolist()
del DF

DF = stockManagerDF[['成长Alpha','重仓股集中度']]
DF['重仓股集中度'] = DF['重仓股集中度']/100
DF = DF.dropna(axis=0, how='any')
X3 = DF['重仓股集中度'].tolist()
Y3 = DF['成长Alpha'].tolist()
del DF

DF = stockManagerDF[['价值Alpha','重仓股集中度']]
DF['重仓股集中度'] = DF['重仓股集中度']/100
DF = DF.dropna(axis=0, how='any')
X4 = DF['重仓股集中度'].tolist()
Y4 = DF['价值Alpha'].tolist()
del DF

fig, axes = plt.subplots(nrows=2,ncols=2)
fig.tight_layout()

axes[0,0].scatter(X1,Y1)
axes[0,0].set_title('重仓股集中度与牛市Alpha关系散点图', fontproperties=prop)
axes[0,0].set_xlabel('重仓股集中度比重（0-1）', fontproperties=prop)
axes[0,0].set_ylabel('牛市Alpha （-1，1）', fontproperties=prop)

axes[1,0].scatter(X2,Y2)
axes[1,0].set_title('重仓股集中度与熊市Alpha关系散点图', fontproperties=prop)
axes[1,0].set_xlabel('重仓股集中度比重（0-1）', fontproperties=prop)
axes[1,0].set_ylabel('熊市Alpha （-1，1）', fontproperties=prop)

axes[0,1].scatter(X3,Y3)
axes[0,1].set_title('重仓股集中度与成长Alpha关系散点图', fontproperties=prop)
axes[0,1].set_xlabel('重仓股集中度比重（0-1）', fontproperties=prop)
axes[0,1].set_ylabel('成长Alpha （-1，1）', fontproperties=prop)


axes[1,1].scatter(X1,Y1)
axes[1,1].set_title('重仓股集中度与价值Alpha关系散点图', fontproperties=prop)
axes[1,1].set_xlabel('重仓股集中度比重（0-1）', fontproperties=prop)
axes[1,1].set_ylabel('价值Alpha （-1，1）', fontproperties=prop)


plt.show()
```
