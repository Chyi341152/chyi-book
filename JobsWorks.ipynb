```python
# 绘制统计直方图
print('统计股票型基金经理总交易日分布和')
DF = stockManagerDF
sum = len(DF['基金经理'])
tradeDays = DF['总交易日'].tolist()
maxtradeDays = max(tradeDays)
# Create a histogram by providing the bin edges (unequally spaced)
bins = [n for n in range(0, maxtradeDays+240) if (n%240) == 0]
orderDict = OrderedDict()
for i in range(len(bins)-1):
    orderDict['{}-{}'.format(bins[i],bins[i+1])] = (len(list(filter(lambda x: x >= bins[i] and x < bins[i+1],tradeDays))))

fig, ax = plt.subplots()
fig.tight_layout()

ind = np.arange(len(orderDict.keys())) # the x locations for the groups
width = 0.35 # the width of the bars: can also be len(x) sequence
rect = ax.bar(ind,[orderDict.get(xtick, 0) for xtick in orderDict.keys()],align='center')
plt.xticks(ind, orderDict.keys(), rotation=30) # Rotating custom tick labels in degress or with keywords
plt.xlabel('TradeDays')
plt.ylabel('num of People')
plt.title('股票型基金经理 交易日统计分布', fontproperties=prop)
def autolabel(rects): # Attach a text label above each bar displaying its height
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width() / 2., 1.005 * height,
                    '%d' % int(height),
                    ha='center', va='bottom')
autolabel(rect)
```
